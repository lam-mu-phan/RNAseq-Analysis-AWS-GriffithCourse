### What my "install_tools.sh" bash script looks like

```bash
#!/bin/bash

set -e  # Exit if any command fails
set -u  # Treat unset variables as errors

echo "Machine architecture:"
uname -m

cd "$RNA_HOME/student_tools/"

echo "Installing HISAT2..."
curl -s https://cloud.biohpc.swmed.edu/index.php/s/oTtGWbWjaxsQ2Ho/download -o hisat2-2.2.1-Linux_x86_64.zip
unzip -o hisat2-2.2.1-Linux_x86_64.zip
cd hisat2-2.2.1
./hisat2 || echo "HISAT2 installed"
cd ..

echo "Installing StringTie..."
wget -q http://ccb.jhu.edu/software/stringtie/dl/stringtie-2.2.1.tar.gz
tar -xzf stringtie-2.2.1.tar.gz
cd stringtie-2.2.1
make release
./stringtie || echo "StringTie installed"
cd ..

# Copy and paste commands to install other tools from https://rnabio.org/module-00-setup/0000/10/01/Installation/

echo "✅ All tools installed."
```

```bash
# To create then execute install_tools.sh, on WSL terminal:
nano install_tools.sh # this will create and open an empty install_tools.sh
# After putting all your commands into bash script, save bash script by ctrl+o, enter, then exit bash script by ctrl+x
# Back to WSL terminal:
chmod +x install_tools.sh # Make it executable
./install_tools.sh # run it, the ./ run the script located in the current directory
```
